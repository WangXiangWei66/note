```python
'''
问题：
当索引为数字时，同时画折线图和柱状图，无法显示折线图
当把索引进行重置后，同时画折线图和柱状图，可以显示折线图
这是为什么？
柱状图与折线图是否存在优先级的问题？
'''

import pandas as pd
import matplotlib.pyplot as plt

index=[398, 1295, 1078, 465, 582, 1311, 1033, 392, 1318, 672]
data=[36, 29, 27, 27, 27, 26, 25, 24, 24, 24]
s=pd.Series(data=data,index=index)
print(s)
s.plot(kind='line',color='r') #kind='line'绘制折线图
s.plot(kind='bar',fontsize=10)  #kind='bar'绘制柱状图,fontsize=10字体大小
```

```python
s1=s.reset_index(drop=True)
s1.plot(kind='line',color='r') #kind='line'绘制折线图
s.plot(kind='bar',fontsize=10)  #kind='bar'绘制柱状图,fontsize=10字体大小
```
